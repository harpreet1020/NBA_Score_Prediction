```python
import numpy as np
import pandas as pd 
from nba_api.stats.endpoints import teamgamelog
```

## Get Team Information 


```python
from nba_api.stats.static import teams

# get_teams returns a list of 30 dictionaries, each an NBA team.
nba_teams = teams.get_teams()

# create pandas DataFrame from nba_teams with columns id, abbreviation
team_id = pd.DataFrame.from_dict(nba_teams)
team_id = team_id[['id']]

custom_headers = {
    'Host': 'stats.nba.com/stats',
    'Connection': 'keep-alive',
    'Cache-Control': 'max-age=0',
    'Upgrade-Insecure-Requests': '1',
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language':'he-IL,he;q=0.8,en-US;q=0.6,en;q=0.4',
}
```

## Get individual Team Game Log

```python

# Pistons 
gamelog = teamgamelog.TeamGameLog(season_all= '2018-19', season_type_all_star='Playoffs', team_id = team_id.loc[28, 'id'], headers=custom_headers)
```

```python
# Get DataFrames 
DET = gamelog.get_data_frames()[0]
DET.head(82)
```

## Get Game Logs for every team 2018-19

```python
df_dict = {}
for team_id in team_id['id']:
    data = teamgamelog.TeamGameLog(season_all='2018-19', season_type_all_star='Playoffs', team_id = team_id, timeout=300, headers=custom_headers).get_data_frames()[0]
    df_dict.update({team_id : data})
```

```python
# Check Type
print(type(df_dict))

# Check length of df_dict to make sure there is data for each team
print(len(df_dict.keys()))
df_dict
```
